```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.5, 
                  shuffle=True, 
                  random_state=0)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c='black', marker='o', s=50)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,
         init='k-means++',
         n_init=10,
         max_iter=300,
         tol=.01,
         random_state=0)
y_km=km.fit_predict(X)
y_km
```

```python
plt.scatter(X[y_km==0,0],
            X[y_km==0,1],
            c='b', marker='x',
            label='0')
plt.scatter(X[y_km==1,0],
            X[y_km==1,1],
            c='r',
            marker='s',
            label='1')
plt.scatter(X[y_km==2,0],
            X[y_km==2,1],
            c='g',
            marker='v',
            label='1')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# kmeans with iris data
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris()
X = iris.data[:, [1, 2]]
y = iris.target
```

```python
plt.scatter(X[:, 0], X[:, 1], c='black', marker='o', s=50)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#Assume there are 2 clusters.
km=KMeans(n_clusters=2,
         init='k-means++',
         n_init=10,
         max_iter=300,
         tol=.01,
         random_state=0)
y_km=km.fit_predict(X)
y_km
```

```python
plt.scatter(X[y_km==1,0],
            X[y_km==1,1],
            c='r',
            marker='s',
            label='1')
plt.scatter(X[y_km==0,0],
            X[y_km==0,1],
            c='g',
            marker='v',
            label='1')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(X[y_km==0,0],
            X[y_km==0,1],
            c='b', marker='x',
            label='0')
plt.scatter(X[y_km==1,0],
            X[y_km==1,1],
            c='r',
            marker='s',
            label='1')
plt.scatter(X[y_km==2,0],
            X[y_km==2,1],
            c='g',
            marker='v',
            label='2')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
