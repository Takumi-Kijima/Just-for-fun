```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Chinese Reastaurant Process

```python
def CRP(num_customers, alpha):

    customer = []
    num_tables = np.zeros(num_customers)
    for i in range(num_customers):
        n = i+1
        #The first customer has only one chioce for choosing a table
        if n == 0:
            customer.append(1)
            num_tables[0] = 1
            choice = 0 #new table is not chosen
        else:
            prob = np.zeros(int(num_tables[i-1])+1)
            for t in range(int(num_tables[i-1])):
                prob[t] =customer.count(t+1) /(alpha + n-1) #the probability that a customer chooses the table (t+1)
            prob[-1]=  alpha / (alpha + n-1) # the probability that a customer chooses the new table
            choice = np.random.multinomial(1, prob, size=1)[0]
            #the cutomer chooses the new table
            if choice[-1] == 1:
                num_tables[i]= num_tables[i-1]+1 # add one more table
                customer.append(int(num_tables[i]))
            else:
                customer.append(int(np.where(np.in1d(choice, 1))[0][0])+1) # the customer n chooses table 
                num_tables[i] = num_tables[i-1]
    return num_tables, customer
```

```python
for i in range(10):
    num_tables, customer = CRP(num_customers=1000, alpha=2)
    plt.plot(list(range(1,1001)), num_tables,color="black")
plt.xlabel('# of customers')
plt.ylabel('# of tables')
plt.title("10 experiments of Chinese Restaurant Process with n=1000 and alpha=2")
plt.show()
```
